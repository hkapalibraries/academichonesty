# Citation Styles
***
### Citation Styles Commonly Used in Arts and Humanities
```{raw} html
<p><span><strong>Citation style</strong> refers to specific rules and guidelines for documenting sources in academic writing.&nbsp;Your instructors may specify citation style for specific assignments or courses.</span></p>

<div style="background:#eeeeee;border:1px solid #cccccc;padding:5px 10px;">
<ul>
	<li><span>Citation Styles (e.g. APA, MLA, or Chicago, etc.)&nbsp;have unique formats for citing various sources.&nbsp;</span></li>
	<li><span>Each citation style provides guidelines on how to format citations for various sources.&nbsp;</span></li>
	<li><span>Each citation style has its own specific formatting guidelines (e.g. hanging indent, line spacing, margin, etc.).</span></li>
	<li><span>Different information formats require different citation formats.</span></li>
	<li><span>Stick to one citation style and use it&nbsp;consistently throughout your work. This means that all your citations, both in-text and in the reference list or bibliography, should follow the same style guideline.&nbsp;</span></li>
</ul>
</div>
```
***

## Citation Styles Guidelines

::::{tab-set}
:::{tab-item} APA 7th
```{raw} html
<div style="overflow: hidden;">
  <img src="https://apastyle.apa.org/Content/Images/megamenu/images@2x/3D-manual-products-menu.png" height="130" width="90" alt="image" style="float: left; margin-right: 10px;">
  <div>
    <a href="http://lib.hkapa.edu/bib/991002046726504326" target="_blank">Publication manual of the American Psychological Association : the official guide to APA style</a><br>
    Author: American Psychological Association<br>
    Call Number: BF76.7 .P83 2020<br>
    Location: Academy Library
  </div>
</div>
<br>
<div style="overflow: hidden;">
  <img src="https://d329ms1y997xa5.cloudfront.net/accounts/206453/images/APA_CHI_7th.jpg" height="130" width="90" alt="image" style="float: left; margin-right: 10px;">
  <div>
    <a href="http://lib.hkapa.edu/bib/991002269622604326" target="_blank">美國心理學會出版手冊 : 論文寫作格式</a><br>
    Author: 鈕文英<br>
    Call Number: BF76.7 .P8312 2022<br>
    Location: Academy Library
  </div>
</div>
```
