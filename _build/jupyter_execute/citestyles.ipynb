```{raw} html
<p><span style="font-size:14px;"><strong>Citation style</strong> refers to specific rules and guidelines for documenting sources in academic writing.&nbsp;<br />
Your instructors may specify citation style for specific assignments or courses.</span></p>

<div style="background:#eeeeee;border:1px solid #cccccc;padding:5px 10px;">
<ul>
	<li><span style="font-size:14px;">Citation Styles (e.g. APA, MLA, or Chicago, etc.)&nbsp;have unique formats for citing various sources.&nbsp;</span></li>
	<li><span style="font-size:14px;">Each citation style provides guidelines on how to format citations for various sources.&nbsp;</span></li>
	<li><span style="font-size:14px;">Each citation style has its own specific formatting guidelines (e.g. hanging indent, line spacing, margin, etc.).</span></li>
	<li><span style="font-size:14px;">Different information formats require different citation formats.</span></li>
	<li><span style="font-size:14px;">Stick to one citation style and use it&nbsp;consistently throughout your work. This means that all your citations, both in-text and in the reference list or bibliography, should follow the same style guideline.&nbsp;</span></li>
</ul>
</div>
```



## **Citation Styles Guidelines**

::::{tab-set}

:::{tab-item} APA 7th
### [RefWorks](http://libproxy.hkapa.edu/login?url=https://refworks.proquest.com/)
A web-based software that helps you to store, organize, and share bibliographic information. There aretwo add-onsdownloadable on different versions of Microsoft Word to facilitate formatting in-text citations, footnotes and bibliographies in any style of your choice, such as APA, MLA, Chicago, and more.\
網絡版書目管理工具，協助你建立及管理個人書目資料庫。撰寫論文時，亦可利用Write-N-Cite或 RefWorks Citation Manager工具，編製符合不同論文格式要求的書目、引用文獻和參考文獻列表。

```{raw} html
<p><strong><span style="font-size:14px;">Citation Examples</span></strong></p>

<div class="panel panel-primary">
<div class="panel-heading">
<h3 class="panel-title"><a aria-controls="collapseEX3" aria-expanded="false" class="btn" data-toggle="collapse" href="#collapseEX3" role="button"><strong>Example 1: Book</strong><span class="glyphicon glyphicon-triangle-bottom"></span></a></h3>
</div>

<div class="panel-body collapse" id="collapseEX3">
<div style="position: relative; width: 100%; height: 0; padding-top: 56.2500%;
 padding-bottom: 0; box-shadow: 0 2px 8px 0 rgba(63,69,81,0.16); margin-top: 1.6em; margin-bottom: 0.9em; overflow: hidden;
 border-radius: 8px; will-change: transform;"><iframe allow="fullscreen" allowfullscreen="allowfullscreen" loading="lazy" src="https://www.canva.com/design/DAGGksu9nCM/JCGOL4--JrR-BrYv7yRP2A/view?embed" style="position: absolute; width: 100%; height: 100%; top: 0; left: 0; border: none; padding: 0;margin: 0;"></iframe></div>
</div>
</div>

<div class="panel panel-primary">
<div class="panel-heading">
<h3 class="panel-title"><a aria-controls="collapseEX4" aria-expanded="false" class="btn" data-toggle="collapse" href="#collapseEX4" role="button"><strong>Example 2: Journal article</strong><span class="glyphicon glyphicon-triangle-bottom"></span></a></h3>
</div>

<div class="panel-body collapse" id="collapseEX4">
<div style="position: relative; width: 100%; height: 0; padding-top: 56.2500%;
 padding-bottom: 0; box-shadow: 0 2px 8px 0 rgba(63,69,81,0.16); margin-top: 1.6em; margin-bottom: 0.9em; overflow: hidden;
 border-radius: 8px; will-change: transform;"><iframe allow="fullscreen" allowfullscreen="allowfullscreen" loading="lazy" src="https://www.canva.com/design/DAGGk1lTnSM/5xf8MpqV3AhDK3HvIgCA7g/view?embed" style="position: absolute; width: 100%; height: 100%; top: 0; left: 0; border: none; padding: 0;margin: 0;"></iframe></div>
</div>
</div>

<div class="panel panel-primary">
<div class="panel-heading">
<h3 class="panel-title"><a aria-controls="collapseEX9" aria-expanded="false" class="btn" data-toggle="collapse" href="#collapseEX9" role="button"><strong>Example 3: Webpage</strong><span class="glyphicon glyphicon-triangle-bottom"></span></a></h3>
</div>

<div class="panel-body collapse" id="collapseEX9">
<div style="position: relative; width: 100%; height: 0; padding-top: 56.2500%;
 padding-bottom: 0; box-shadow: 0 2px 8px 0 rgba(63,69,81,0.16); margin-top: 1.6em; margin-bottom: 0.9em; overflow: hidden;
 border-radius: 8px; will-change: transform;"><iframe allow="fullscreen" allowfullscreen="allowfullscreen" loading="lazy" src="https://www.canva.com/design/DAGYUFwY35M/RUvz_zfg4-E_rw1NdYznYw/view?embed" style="position: absolute; width: 100%; height: 100%; top: 0; left: 0; border: none; padding: 0;margin: 0;"></iframe></div>
</div>
</div>

<div class="panel panel-primary">
<div class="panel-heading">
<h3 class="panel-title"><a aria-controls="collapseEX10" aria-expanded="false" class="btn" data-toggle="collapse" href="#collapseEX10" role="button"><strong>Example 4: Online Video</strong><span class="glyphicon glyphicon-triangle-bottom"></span></a></h3>
</div>

<div class="panel-body collapse" id="collapseEX10">
<div style="position: relative; width: 100%; height: 0; padding-top: 56.2500%;
 padding-bottom: 0; box-shadow: 0 2px 8px 0 rgba(63,69,81,0.16); margin-top: 1.6em; margin-bottom: 0.9em; overflow: hidden;
 border-radius: 8px; will-change: transform;"><iframe allow="fullscreen" allowfullscreen="allowfullscreen" loading="lazy" src="https://www.canva.com/design/DAGYTWGt-BA/ZbITQVfzQjUmVgyH3Sm_Xw/view?embed" style="position: absolute; width: 100%; height: 100%; top: 0; left: 0; border: none; padding: 0;margin: 0;"></iframe></div>
</div>
</div>

<p>Source:&nbsp; <a href="https://apastyle.apa.org/style-grammar-guidelines/references/examples#textual-works" target="_blank">APA Style-&nbsp;Reference Examples</a></p>

<hr />
<p><span style="font-size:14px;"><strong>Learn more:</strong></span></p>

<div>
<div style="float:left; margin-right: 10px;"><a href="https://academicwriter-apa-org.libproxy.hkapa.edu/learndisplay/TUT-11"><img loading="lazy" src="https://academicwriter.apa.org/learnasset/thumbnails/small/TUT/TUT-11.jpg" style="vertical-align: middle;" /></a></div>

<div>
<p><a href="https://academicwriter-apa-org.libproxy.hkapa.edu/learndisplay/TUT-11" style="text-decoration: none;"><b>Basics of APA Style</b></a></p>

<p><meta name='EZproxyRewriting' content='disable'>Learn the basics of seventh edition APA Style, including paper elements, format, and organization; academic writing style; grammar and usage; bias-free language; mechanics of style; tables and figures; in-text citations, paraphrasing, and quotations; and reference list format and order.<meta name='EZproxyRewriting' content='enable'></p>

<p style="margin-bottom: 0;">Academic Writer</p>

<p style="margin-top:0;">&copy; 2024 American Psychological Association.</p>
</div>

<p style="clear:both; margin:0; padding:0;">&nbsp;</p>
</div>
```
